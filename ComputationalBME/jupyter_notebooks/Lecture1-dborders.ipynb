```python
import pandas as pd

csvfl=pd.read_csv("../ds4bme_intro/data/kirby21.csv")
```

```python
#EXERCISE 1
TL=list(csvfl[csvfl.roi=="Telencephalon_L"].volume)
TR=list(csvfl[csvfl.roi=="Telencephalon_R"].volume)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.xlabel("TL")
plt.ylabel("TR")
plt.title("TR vs. TL - my own functions")
plt.scatter(TL, TR)
    
#EXERCISE 2

import numpy as np

#Predicting with TL
s1=float(sum(list(map(lambda x: x[0]*x[1], zip(TL, TR)))))/sum(list(map(lambda x: x**2, TL)))

#Plot the TL line
x_l=np.linspace(min(TL), max(TL), 1000)
y_l=x_l*s1
plt.plot(x_l, y_l)

#Predicting with TR
s2=float(sum(list(map(lambda x: x[0]*x[1], zip(TR, TL)))))/sum(list(map(lambda x: x**2, TR)))

#Plot the TR line
x_r=np.linspace(min(TR), max(TR), 1000)
y_r=x_r*s2
plt.plot(y_r, x_r)

#Note, the regressions are almost identicle so the lines overlap

plt.legend(["Pred w/ TL", "Pred w/ TR"])
plt.show()


#EXERCISE 2 continued
from scipy import stats

plt.figure()
plt.xlabel("TL")
plt.ylabel("TR")
plt.title("TR vs. TL - built in functions")

for x, y in zip(TL, TR):
    plt.plot(x, y, "o")
    
#using TL as x
s1, i1, r1, p1, se1=stats.linregress(TL,TR)

#Plot the TL line
x_l=np.linspace(min(TL), max(TL), 1000)
y_l=x_l*s1+i1
plt.plot(x_l, y_l)

#using TR as x
s2, i2, r2, p2, se2=stats.linregress(TR,TL)

#Plot the TR line
x_r=np.linspace(min(TR), max(TR), 1000)
y_r=x_r*s2+i2
plt.plot(y_r, x_r)

plt.legend(["Pred w/ TL", "Pred w/ TR"])
plt.show()

#Note, the slopes are a bit diffrent because my method goes through the origin 
#and scipy does not go through the origin.
```

$$y=bx$$ 
<center>but then replace:</center> 
$$x --> cx$$

$$B=\sum y_i x_i / \sum x_i^2$$

<center>becomes:</center>
$$B=\sum y_i c x_i / \sum x_i^2 c^2$$

$$B=\sum y_i x_i / c\sum x_i^2$$

<center>So the regression slope will be scaled by a factor of: $$1/c$$<center/>

This makes sense because the function will essentially be stretched along the x axis so if c>1, then the slope will need to get smaller.
